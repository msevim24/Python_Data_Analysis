# Missing Values Imputation

```python
# null değerlerin sayısını ve yüzdesini hesaplamak ve bunları bir DataFrame'de sunmak için tanımladığımız fonksiyon


def analyze_missing_values(data):
    missing_df = pd.DataFrame({
        "count":
        data.isnull().sum(),
        "percentage":
        round(data.isnull().mean() * 100, 2)
    })
    missing_df = missing_df[missing_df['count'] != 0].sort_values(
        by='count', ascending=True)
    return missing_df
```

```python
analyze_missing_values(data)
```

```python
# 'gross' ve 'budget' sütunlarında çok sayıda NaN değeri bulunduğu için, bu sütunlardaki NaN değerlerine
# sahip tüm satırları düşürüyoruz.

# Bunun için NumPy'ın 'isnan' fonksiyonunu bir negasyon (negasyon, bir ifadenin veya bir durumun tersini
# ifade etmek anlamına gelir) '~' (tilde işareti) ile birlikte kullanıyoruz.

data = data[~np.isnan(data['gross']) & ~np.isnan(data['budget'])]

analyze_missing_values(data)
```

```python
# Tüm satırlardaki toplam Null değer sayısına bakar, sadece Null değerler toplamı ikiden küçük olan satırlar korunur

data = data[data.isnull().sum(axis=1) <= 2]

analyze_missing_values(data)
```

```python
# Eksik değerleri dolduralım

# Sayısal sütunlar için ortalama değer kullanıyoruz
data['num_critic_for_reviews'].fillna(data['num_critic_for_reviews'].mean(),
                                      inplace=True)
data['duration'].fillna(data['duration'].mean(), inplace=True)

# Kategorik sütun için mod kullanıyoruz
data['language'].fillna(data['language'].mode()[0], inplace=True)

# Oyuncu isimleri gibi istatistiksel değerlerle doldurulamayacak kategorik veriler için "Unknown Actor" ifadesini kullanıyoruz
data['actor_2_name'].fillna('Unknown Actor', inplace=True)
data['actor_3_name'].fillna('Unknown Actor', inplace=True)

analyze_missing_values(data)
```

# Feature Engineering

```python
# Brüt kazanç ve bütçeyi $'dan Milyon $'a dönüştürelim ki analizimiz daha kolay olsun

data['gross'] = data['gross'] / 1000000
data['budget'] = data['budget'] / 1000000
```

```python
# Kar sütunu oluşturalım, Brüt kazanç (gross) ve Bütçe (budget) kullanılarak

data['Profit'] = data['gross'] - data['budget']
```

```python
# Yukarıdaki çıktıya bakarak, 3,500 film arasından sadece 150-200 filmin farklı dillerde olduğunu kolayca gözlemleyebiliriz

# Bu nedenle sadece İngilizce ve Yabancı olmak üzere iki dil kategorisi belirlemek daha iyidir


def language(x):
    if x == 'English':
        return 'English'
    else:
        return 'Foreign'
```

```python
# Fonksiyonu 'language' sütununa uygulayalım

data['language'] = data['language'].apply(language)
```

```python
# Değerleri tekrar kontrol edelim

data['language'].value_counts()
```

```python
# Filmlerin süreleri çok fazla değişkenlik göstermiyor ama bildiğimiz üzere çoğu kullanıcı ya uzun ya da kısa süreli
# filmleri tercih ediyor.
# Filmleri 'Kısa' ve 'Uzun' olarak iki kategoriye ayırabiliriz.


# Filmlerin süresini kategorize edecek bir fonksiyon tanımlayalım

def duration(x):
    if x <= 120:
        return 'Short'
    else:
        return 'Long'
```

```python
# Bu fonksiyonu 'duration' sütununa uygulayalım

data['duration'] = data['duration'].apply(duration)
```

```python
# 'duration' sütunundaki değerleri kontrol edelim

data['duration'].value_counts()
```

```python
# 'genres' sütunundaki değerleri kontrol edelim

data['genres'].value_counts()
```

```python
# 'genres' sütunundaki ilk film için türlerin nasıl ayrıldığına bakalım

data['genres'].str.split('|')[0]
```

```python
# Yukarıdaki çıktıdan görebileceğimiz üzere, çoğu film birçok farklı türe sahip
# Ayrıca bir film birden fazla türe sahip olabileceği için, dört ana türü tutmak yeterli olacaktır

# Filmlerin türlerini ayrı bir sütuna ayıralım
data['Moviegenres'] = data['genres'].str.split('|')
data['Genre1'] = data['Moviegenres'].apply(lambda x: x[0])

# Bazı filmler sadece bir türe sahip. Bu durumlarda 'Genre2' için de aynı türü atayalım
data['Genre2'] = data['Moviegenres'].apply(lambda x: x[1]
                                           if len(x) > 1 else x[0])
data['Genre3'] = data['Moviegenres'].apply(lambda x: x[2]
                                           if len(x) > 2 else x[0])
data['Genre4'] = data['Moviegenres'].apply(lambda x: x[3]
                                           if len(x) > 3 else x[0])
```

## Analysis of Movies by a Specific Actor

```python
# Belirli Bir Aktöre Ait Filmlerin Detaylı Analizi


def query_actors(x):
    # Belirtilen aktörün birinci, ikinci veya üçüncü aktör olarak listelendiği filmleri birleştir
    a = data[data['actor_1_name'] == x]
    b = data[data['actor_2_name'] == x]
    c = data[data['actor_3_name'] == x]
    y = pd.concat([a, b, c])

    y = y[[
        'movie_title',
        'budget',
        'gross',
        'title_year',
        'genres',
        'language',
        'imdb_score',
    ]]

    y = y.sort_values(by='imdb_score', ascending=False)
    return y
```

```python
query_actors('Christian Bale')
```

## Analysis of Performance in Actor-Based Movies

```python
def actors_report(x):
    a = data[data['actor_1_name'] == x]
    b = data[data['actor_2_name'] == x]
    c = data[data['actor_3_name'] == x]
    y = pd.concat([a, b, c])

    print("Time:", y['title_year'].min(), y['title_year'].max())
    print("Max Gross : {0:.2f} Millions".format(y['gross'].max()))
    print("Avg Gross : {0:.2f} Millions".format(y['gross'].mean()))
    print("Min Gross : {0:.2f} Millions".format(y['gross'].min()))
    print("Number of 100 Million Movies :", y[y['gross'] > 100].shape[0])
    print("Avg IMDB Score : {0:.2f}".format(y['imdb_score'].mean()))
    print("Most Common Genres:\n", y['Genre1'].value_counts().head())


actors_report('Christian Bale')
```

## Analysis of the Number of Reviews Evaluating Actors by Critics

```python
def critically_acclaimed_actors(m):
    frames = [
        data[data['actor_1_name'] == m], data[data['actor_2_name'] == m],
        data[data['actor_3_name'] == m]
    ]
    combined = pd.concat(frames)
    total_reviews = combined['num_critic_for_reviews'].sum()
    # NaN değerlerini kontrol et ve 0'a dönüştür
    return int(total_reviews) if pd.notnull(total_reviews) else 0
```

```python
# Aktörler için eleştirmen incelemelerinin sayısını yazdırma

print("Number of Critics Reviews for Al Pacino")
display(critically_acclaimed_actors('Al Pacino'))
```

```python
print("Number of Critics Reviews for Meryl Streep")
display(critically_acclaimed_actors('Meryl Streep'))
```

**mlxtend (Machine Learning Extensions):**

- Makine öğrenimi, veri bilimi ve veri madenciliği uygulamaları için kullanışlı araçlar ve algoritmalar sunan bir Python kütüphanesidir. 

- mlxtend.preprocessing ise, mlxtend kütüphanesinin ön işleme modülüdür ve veri ön işleme işlemleri için çeşitli sınıflar ve fonksiyonlar içerir.

- Bu modül, veri setlerinin makine öğrenimi algoritmalarına uygun hale getirilmesi için gerekli olan bir dizi ön işleme işlemi gerçekleştirmeyi kolaylaştırır. 

- İşlemler arasında özellik ölçeklendirme (feature scaling), kategorik verinin sayısal veriye dönüştürülmesi, veri setlerindeki işlemlerin ikili formatlara dönüştürülmesi gibi işlemler bulunur.

**TransactionEncoder:**
- TransactionEncoder, bir dizi işlem (örneğin, alışveriş sepeti işlemleri veya burada olduğu gibi film türleri listesi) içindeki öğeleri ikili formatına dönüştürmek için kullanılan bir sınıftır. 

- Her bir işlemde öğenin var olup olmadığını gösteren bir ikili matris (True/False veya 1/0) oluşturur. Bu, sık öğe seti madenciliği gibi analizler için öğelerin varlığını kolayca belirlemeyi sağlar.

- Bu özel durumda, TransactionEncoder film türlerini içeren listeleri alır ve her bir türün filmlerde var olup olmadığını gösteren ikili bir matris oluşturur. 

- Sonuç olarak, her bir sütun bir film türünü temsil eder ve her satır, o türün ilgili filmde olup olmadığını (1 veya 0) gösterir. 

- Bu dönüşüm, türlerin analizi ve modelleme için veri setini hazır hale getirir.

https://rasbt.github.io/mlxtend/user_guide/preprocessing/TransactionEncoder/




```python
data['genres']
```

```python
# 'genres' sütunundaki türleri '|' karakterine göre ayır ve bir liste haline getir

x = data['genres'].str.split('|')
```

```python
# TransactionEncoder nesnesi oluştur

te = TransactionEncoder()
```

```python
# Oluşturulan listeyi TransactionEncoder ile fit ve transform işlemine tabi tut

x = te.fit_transform(x)
```

```python
# Sonucu bir pandas DataFrame'ine dönüştür ve sütun adlarını TransactionEncoder'dan al

x = pd.DataFrame(x, columns=te.columns_)
```

```python
x.head()
```

```python
# Veriyi boolean'dan tamsayıya çevirelim ki hesaplamalar yapabilelim

genres = x.astype('int')

genres.head()
```

```python
# Şimdi, veriyi daha iyi anlayabilmek için ilk sütuna film başlıklarını ekleyelim

genres.insert(0, 'movie_title', data['movie_title'])

genres.head()
```

```python
# Bu film başlıklarını verinin indeksi yapalım

genres = genres.set_index('movie_title')

genres.head()
```

```python
# Benzer türleri bulmak için bir öneri motoru yapalım

# Hedef tür ile diğer tüm türler arasındaki korelasyonu hesaplayalım


def recommendation_genres(gen):
    gen = genres[gen]
    similar_genres = genres.corrwith(gen)
    similar_genres = similar_genres.sort_values(ascending=False)
    similar_genres = similar_genres.iloc[1:]
    return similar_genres.head(3)
```

**corrwith Fonksiyonu**

- Pandas kütüphanesindeki DataFrame objeleri için kullanılır ve bir DataFrame ile diğer bir Series veya DataFrame arasındaki sütun bazında korelasyonları hesaplar. 

- Bu fonksiyon, iki veri seti arasındaki ilişkiyi ölçmek için Pearson korelasyon katsayısını kullanır. 

- Brada hedef tür ile diğer tüm türler arasındaki korelasyonu hesaplar.

- Benzer türleri belirlemek için etkili bir yöntemdir çünkü yüksek korelasyon değerleri, türler arasında güçlü bir ilişki olduğunu gösterir. Bu şekilde, kullanıcıların ilgi alanlarına uygun yeni türler keşfetmeleri sağlanır.

```python
recommendation_genres('Comedy')
```

# Recommending similar Movies

```python
# 'genres' DataFrame'ini transpoze ederek yeni bir yapı oluşturuyoruz.

x = genres.transpose()
x.head()
```

```python
# Belirli bir film için benzer filmleri bulmak üzere fonksiyon


def recommendation_movie(movie):
    movie = x[movie + '\xa0']
    similar_movies = x.corrwith(movie)
    similar_movies = similar_movies.sort_values(ascending=False)
    similar_movies = similar_movies.iloc[1:]
    return similar_movies.head(20)
```

**Non-breaking Space ve Veri İşleme**

- Veri işleme sırasında, veri setindeki film adlarının sonunda bazen istenmeyen boşluklar veya özel karakterler bulunabilir. 

- Bu, veri toplama ve temizleme aşamalarında, veri kaynaklarının tutarsızlıklarından kaynaklanabilir. 

- **movie+'\xa0'** ifadesi, veri setindeki film adlarının sonunda bir non-breaking space karakteri (\xa0) olabileceğini varsayar ve bu karakteri programatik olarak ekleyerek doğru eşleşmeyi sağlamaya çalışır.
